```python
import sqlite3
from sqlite3 import Error
from pathlib import Path
import subprocess
import pandas as pd
from pandas import DataFrame as df
import datetime
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import sklearn
```

```python
def create_connection(db_file):
    """ create a database connection to a SQLite database """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
        
    return conn
```

```python
def get_data(conn, sql):

    try:
        c = conn.cursor()
        c.execute(sql)
        return c.fetchall()
    except Error as e:
        print(e)
```

```python
Customers_RFM_Query = """
SELECT MemberId, MAX(Date) AS Last_Purchase_Date, COUNT(DISTINCT Date) AS Frequency, COUNT() AS MonetaryValue
FROM Purchases
GROUP BY MemberId """
```

```python
con = create_connection(r"Grocery_Store")
Customers_RFM_Unscored = pd.read_sql(Customers_RFM_Query, con)
Customers_RFM_Unscored = Customers_RFM_Unscored.set_index("MemberId")
Customers_RFM_Unscored.head(5)
```

```python
Customers_RFM_Unscored['Last_Purchase_Date'] = pd.to_datetime(Customers_RFM_Unscored['Last_Purchase_Date'], format= '%Y-%m-%d')
Customers_RFM_Unscored['Last_Purchase_Date'] = Customers_RFM_Unscored['Last_Purchase_Date'].dt.date
```

```python
last_day = Customers_RFM_Unscored['Last_Purchase_Date'].max() + datetime.timedelta(days=1)
print(last_day)
```

```python
Customers_RFM_1 = Customers_RFM_Unscored.copy()
Customers_RFM_1.columns = ['Recency', 'Frequency', 'Monetary']
Customers_RFM_1['Recency'] = (last_day - Customers_RFM_1['Recency']).dt.days
Customers_RFM_1.head(10)
```

```python
sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(customers_RFM_Normalised_df)
    sse[k] = kmeans.inertia_ # SSE to closest cluster centroidplt.title('The Elbow Method')
    
plt.xlabel('k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```python
sseSlope = {}
```

```python
for i in range(1,10):
    sseSlope[i] = sse[i+1] - sse[i]
```

```python
plt.xlabel('k')
plt.ylabel('SSE Slope')
sns.pointplot(x=list(sseSlope.keys()), y=list(sseSlope.values()))
plt.show()
```

```python
model = KMeans(n_clusters=6, random_state=42)
model.fit(customers_RFM_Normalised_df)
model.labels_.shape
customers_RFM["Cluster"] = model.labels_
customers_RFM.groupby('Cluster').agg({
'Recency':'mean',
'Frequency':'mean',
'Monetary':['mean', 'count']})
```

```python
f, ax = plt.subplots(figsize=(25, 5))
ax = sns.countplot(x="Cluster", data=customers_RFM)
```

```python
fig_data_normCust = pd.DataFrame(customers_RFM_Normalised_df, columns=['Recency', 'Frequency', 'Monetary'])
fig_data_normCust['ID'] = customers_RFM_Normalised_df.index
fig_data_normCust['Cluster'] = model.labels_
# Melt The Data
df_nor_melt = pd.melt(fig_data_normCust.reset_index(),
                      id_vars=['ID', 'Cluster'],
                      value_vars=['Recency','Frequency','Monetary'],
                      var_name='Attribute',
                      value_name='Value')
df_nor_melt.head()
```
